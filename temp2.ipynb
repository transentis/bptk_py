```python
from BPTK_Py import Model
import BPTK_Py
from BPTK_Py import sd_functions as sd


class Floor(sd.Function):
    """
    Generic SD DSL function.
    """

    def __init__(self, element):
        self.element = element
    

    def term(self, time="t"):
        return "math.floor({})".format(self.element)



model = Model(starttime=0.0, stoptime=20.0, dt=1.0, name="Test Model")

model.points["dynamic_customer_order"] = [(1.0, 100.0), (2.0, 400.0), (6.0, 400.0), (7.0, 200.0)]
flow = model.flow("flow")
flow.equation = sd.lookup(sd.time(), "dynamic_customer_order")
# flow.plot()

test_c = model.converter("test_c")
test_c.equation = 2.2
test_d = model.converter("test_d")
test_d.equation = Floor(test_c)

test_d.plot()
```
